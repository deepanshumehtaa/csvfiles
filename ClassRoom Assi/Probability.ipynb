# Normalizing

```
# np.random.normal(size=4)     # genrating 4 random no. from normal distribution

arr = np.random.randint(low=-8, high=10, size=15)  # genrate 4 nos.
print(arr)

def plott(a):
    
# The trick is i can't use scatter plot to plot the points above but i need to use line plot wisely
    plt.plot(a, '*')     # Stars

# use to plot axis
    plt.axhline(0, color='black')  # horizontal
    plt.axvline(0, color='black')  # vertical

    plt.show()

plott(arr)
```

## find mean and shift that to origin

```
dif = int(np.mean(arr))

arr1 = arr - dif
arr1
```
