# Train and Testing
logistic regression

```
# Train

x = file_data.drop(['Survived'], axis=1)
y = file_data["Survived"]
```

```
from sklearn.model_selection import train_test_split
# train_test_split, it split data into train and test randomly
# and save them X_train, X_test, Y_train, Y_test

# same as
# from sklearn.cross_validation import train_test_split
```

```
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.3, random_state= 42)
# Train pehle ati hai
# dividing test_size = 0.3 and 0.7, random_state= anything
# logistic regression follow trend of linear regression
```

```
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
# logmodel is object which v r going to use
```

```

lr.fit(X_train, Y_train)
```

```
prediction = lr.predict(X_test)
print(prediction)
```

```
from sklearn.metrics import classification_report
classification_report(Y_test, prediction)
```
